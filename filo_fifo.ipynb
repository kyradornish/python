## FiLo

```python
class Item():
    def __init__(self, name):
        self.name = name
        
    def getName(self):
        return self.name

class Stackz():
    
    def __init__(self):
        self.zl = []
    
    def add(self, item):
        self.zl.insert(0, item)
        
    def remove(self):
        self.zl.pop(0)
        
    def __iter__(self):
        yield from self.zl
    
    def printList(self):
        for i in self.zl:
            print("{}".format(i.getName()))

    

coin = Item("coin")
bag = Item("bag")
dog = Item("dog")
cat = Item("cat")
inventory = Stackz()
inventory.add(coin)
inventory.add(bag)
inventory.add(dog)
inventory.add(cat)
inventory.remove()
inventory.remove()
inventory.printList()
```

## FiFo

```python
class Item2():
    def __init__(self, name):
        self.name = name
        
    def getName2(self):
        return self.name

class Queue():
    
    def __init__(self):
        self.zl2 = []
    
    def add(self, item):
        self.zl2.append(item)
        
    def remove(self):
        del self.zl2[0]
        
    def __iter__(self):
        yield from self.zl2
    
    def printList(self):
        for i in self.zl2:
            print("{}".format(i.getName2()))
    
    
coin = Item2("coin")
bag = Item2("bag")
dog = Item2("dog")
cat = Item2("cat")
inventory2 = Queue()
inventory2.add(coin)
inventory2.add(bag)
inventory2.add(cat)
inventory2.remove()
inventory2.printList()
```
