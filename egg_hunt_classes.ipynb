```python
from random import randint
from IPython.display import clear_output

class Board():
    grid = []
    def __init__(self, rows, cols):
        self.rows = rows
        self.cols = cols
        
    def print_board(self):
        for x in range(self.rows):
            self.grid.append(["__"]*self.cols)
        for i in self.grid:
            print (" ".join(i))
        
    def getGridLength(self):
        return len(self.grid)
    

class Thing():
    def __init__(self):
        self.coords = [randint(0, length - 1), randint(0, length - 1)]

    
class Person():
    def __init__(self):
        self.coords = [0,0]
        
    def movePlayer(self, move):
        if move == 'left':
            self.coords[0] -= 1
        if move == 'right':
            self.coords[0] += 1
        if move == 'up':
            self.coords[1] -= 1
        if move == 'down':
            self.coords[1] += 1
        return self.coords
    
class Door():
    def __init__(self):
        self.coords = [0,0]

#Print Instructions
print("Welcome to The Easter Egg Hunt!\n-->Collect the 3 hidden eggs and exit through the door.\n-->You must first find the basket to collect eggs.\n-->Don't run into the hidden monster!")
print("Enter 'QUIT' to quit.")

#Set board to 5x5
things = ["basket", "monster", "egg1", "egg2", "egg3"]
board = Board(5, 5) 
#Print board to console
board.print_board()

#Define var with length of board
length = board.getGridLength()

#Setting random locations to each Thing
monster = Thing()
egg1 = Thing()
egg2 = Thing()
egg3 = Thing()
basket = Thing()
player = Person()
door = Door()

print(monster.coords)
print(egg1.coords)
print(egg2.coords)
print(egg3.coords)
print(basket.coords)

eggs = 0
eggbasket = 0
found = []

while True:
    print("You're currently in room {}".format(player.coords))
    move = input("> ")
    player.movePlayer(move)
    
    if move == "QUIT":
        break
        
    if player.coords == monster.coords:
        print("You ran into a monster and died! Game over.")
        break
    if player.coords == egg1.coords: 
        if eggbasket > 0:
            found.append("egg1")
            eggs += 1
            print("Congratulations you found an egg! So far, you have found {}".format(found))
        else:
            print("You must first find the basket before you can collect eggs.")
    if player.coords == egg2.coords: 
        if eggbasket > 0:
            found.append("egg2")
            eggs += 1
            print("Congratulations you found an egg! So far, you have found {}".format(found))
        else:
            print("You must first find the basket before you can collect eggs.")
    if player.coords == egg3.coords: 
        if eggbasket > 0:
            found.append("egg3")
            eggs += 1
            print("Congratulations you found an egg! So far, you have found {}".format(found))
        else:
            print("You must first find the basket before you can collect eggs.")
    if player.coords == basket.coords:
        found.append("basket")
        eggbasket += 1
        print("Congrats you have found the basket. So far, you have found {}".format(found))
    if player.coords == door.coords:
        if eggs >= 3:
            clear_output()
            print("CONGATS you won!")
            break
```
