```python
from IPython.display import clear_output

#Show instructions
def instructions():
    print("Enter 'Show' to see a list of your items in your cart.") 
    print("Enter 'Help' to see the instructions.")
    print("Enter 'Clear Cart' to clear your cart.")
    print("Enter 'Delete Item' to delete an item.")
    print("Enter 'Done' proceed.")

#Add items to the list
def addtolist(var):
    slist.append(var)

#Clear the list
def clear_list(slist):
    del slist[:]

#Input an item to delete - deletes specific item 
def delete_item():
    item_to_remove = input("Which item would you like to delete?")
    
    temp = -100
    try:
        temp = int(item_to_remove)
        if temp >= 0 and temp <= len(slist):
            del slist[temp]
        clear_output()
        print("{} was deleted from the list".format(item_to_remove))
    except:
        temp = -100
        slist.remove(item_to_remove)
        clear_output()
        print("{} was deleted from the list".format(item_to_remove))
    
    
slist = []

instructions()

while True:
    var = input("Type something: ")
    if var == 'Help':
        clear_output()
        instructions()
    elif var == 'Show':
        clear_output()
        for items in range(len(slist)):
            print("{}: {}".format(items+1, slist[items]))
    elif var == 'Clear Cart':
        clear_output()
        clear_list(slist)
        print("Cart is Empty")
    elif var == 'Delete Item':
        clear_output()
        for items in range(len(slist)):
            print("{}: {}".format(items+1, slist[items]))
        delete_item()
    elif var == 'Done':
        clear_output()
        for items in slist:
            print(items)
        break
    else:
        clear_output()
        addtolist(var)
```
