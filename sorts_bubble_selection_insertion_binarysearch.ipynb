## Bubble Sort

```python
arr1 = [5,1,25,18]
```

```python
def bubbleSort(arr1):
    
    for i in range(len(arr1)):
        for j in range(len(arr1)-1):
            if arr1[j] > arr1[j+1]:
                larger_num = arr1[j]
                arr1[j] = arr1[j+1]
                arr1[j+1] = larger_num
    print(arr1)
    
bubbleSort(arr1)
```

```python
def bubbleSort(arr1):
    
    for i in range(len(arr1)):
        for j in range(len(arr1)-1):
            if arr1[j] > arr1[j+1]:
                arr1[j], arr1[j+1] = arr1[j+1], arr1[j]
    print(arr1)
    
bubbleSort(arr1)
```

## Selection Sort

```python
import random
```

```python
li = [random.randint(0, 50) for i in range(10)]
li
```

```python
def selcSort(Li):               
    for i in range(0, len(Li)-1):
        print(Li)
        for j in range(i+1, len(Li)):
            if Li[j] <= Li[i]:
                Li[j], Li[i] = Li[i], Li[j]
    return Li
            
selcSort(li)
```

## Insertion Sort

```python
li2 = [random.randint(0, 50) for i in range(10)]
li2
```

```python
def insertSort(L2):
    for index in range(1, len(L2)):
        value = L2[index]
        i = index-1
        while i >= 0:
            if value < L2[i]:
                L2[i+1] = L2[i]
                L2[i] = value
                i -=1
            else:
                break
    return li2

insertSort(li2)
```

## Binary Search

```python
def binarySearch(item, Li):
    Li_copy = sorted(Li)
    print(item)
    while True:
        print(Li_copy)
        mid = int(len(Li_copy) / 2)
        if item == Li_copy[mid]:
            return True
        elif item > Li_copy[mid]:
            del Li_copy[:mid] 
        else:
            del Li_copy[mid:]
        if mid == 0:
            break
    return False

litest = [random.randint(0, 10) for i in range(10)]
binarySearch(random.randint(0,10), litest)
```
